```python
import json
import os
from bs4 import BeautifulSoup as BS
import re
import requests
import pandas as pd
```

```python
main_url = "https://finance.naver.com/item/news_news.nhn?code="
```

```python
com_code = "005930" # KRX 종목코드
```

```python
def cleartext(text):
    text = text.replace("\n","").replace("\t","").replace("\r","").replace("r_start","").replace("r_end","").strip(" ")
    text = text.replace("// flash 오류를 우회하기 위한 함수 추가function _flash_removeCallback() {}","")
    text = text.replace("\xa0"," ")
    return text
```

```python
news_file = pd.DataFrame({"date":[], "body":[]}, columns = ["date", "body"])
```

```python
%%time

for i in range(1,1000):
    
    url_full = main_url + com_code +"&page="+str(i)+"&sm=title_entity_id.basic&clusterId="
    
    r = requests.get(url_full)

    soup = BS(r.text, "html")
    
    news_title = soup.select("a.tit")
    
    news_date = soup.select("td.date")
    
    print("{} pages crawling... ".format(i))
    
    
    for j in range(len(news_title)):
        
        
        news_title_str = str(news_title[j])
        news_date_str = str(news_date[j])
        
                
        art_start = news_title_str.index("article_id=") + 11
        art_end = art_start + 10
        
        art_id = str(news_title_str[art_start:art_end])
        
                        
        offi_start = news_title_str.index("office_id=") + 10
        offi_end = offi_start + 3
        
        offi_id = str(news_title_str[offi_start:offi_end])
                
        art_sub = "https://finance.naver.com/item/news_read.nhn?article_id={}&office_id={}&code={}&page=1&sm=title_entity_id.basic".format(art_id, offi_id,com_code)
        
                
        news_date_start = news_date_str.index(">")+1
        news_date_end = news_date_start+11
        
        news_date_final = news_date_str[news_date_start:news_date_end]
                
        
        #print(art_id)
        #print(offi_id)
        #print(news_date_final)
        #print(art_sub)
        
        arti_r = requests.get(art_sub)
        arti_html = BS(arti_r.text, "html.parser")        
        arti_html_str = str(arti_html.select('#news_read'))
        
                
        try:
        
            links_foot = arti_html_str.index('<div class="link_news">')
            arti_html_mid = arti_html_str[:links_foot]
            
        except:
            pass
        
                
        arti_html_final = re.sub('<[A-Za-z\/][^>]*>', '', arti_html_mid)
        arti_html_final = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』;★▲◇◆\\‘|\(\)\[\]\<\>`\'…》]', '', arti_html_final)
        
        try:
        
            arti_html_final = arti_html_final[:arti_html_final.index("▶")]
            
        except:
            
            pass
        
        #print(cleartext(arti_html_final))
        
        #print(arti_html_final)
        news_file.ix[art_sub] = [news_date_final, arti_html_final]
        
                
        #print("-----------------------------------------")
```
